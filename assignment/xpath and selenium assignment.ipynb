```python
from selenium import webdriver
```

```python
jumia = 'https://www.jumia.com.gh/phones-tablets/'
drivers = webdriver.Chrome()
drivers.get(jumia)
```

```python
product = [jumpt.text for jumpt in drivers.find_elements_by_xpath("//div[@class= 'name']")]
product
len(product)
```

```python
prices1 = [price.text for price in drivers.find_elements_by_xpath("//div[@class= 'prc']")]
prices1
len(prices1)
```

```python
jptabs_link1 = []
jptabs_link = drivers.find_elements_by_xpath("//*[@id='jm']/main/div[2]/div[3]/section/div[1]/article/a")


for i in jptabs_link:
    jptabs_links = print(i.get_attribute('href'))
    jptabs_link1.append(i.get_attribute('href'))
print(jptabs_link1)
len(jptabs_link1)
```

```python
import pandas as pd
df = pd.DataFrame({'Products': product[:48], 'Prices':prices1[:48], 'Link':jptabs_link1[:48] })
df
```
